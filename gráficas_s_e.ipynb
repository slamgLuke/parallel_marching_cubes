```python


# Parámetros
threads = np.array([2, 4, 8, 16, 32, 64, 128])
Ns = np.array([32, 64, 128, 256, 512, 1024])
d = 8  # deph
#
C = 0.001
D = 0.0001

def T_parallel(N, p, d, C, D):
    epsilon = 10 / N
    work = (N**3) / (p )
    critical = C * (N**3) * np.log2(N) / (p )
    tasks = D * (8**d) * np.log2(N)
    return work + critical + tasks


def speedup(N, p, d, C, D):
    T1 = (N**3)
    Tp = T_parallel(N, p, d, C, D)
    return T1 / Tp


plt.figure(figsize=(12, 7))
colors = itertools.cycle(plt.cm.tab10.colors)
for N, color in zip(Ns, colors):
    speedups = [speedup(N, p, d, C, D) for p in threads]
    plt.plot(threads, speedups, 'o-', label=f'N={N}', color=color, markersize=8)


plt.xscale('log', base=2)
plt.xlabel('Número de Threads (p)')
plt.ylabel('Speedup $S_p$')
plt.title('Speedup en Weak scaling')
plt.legend(title='Tamaño del cubo', bbox_to_anchor=(1.05, 1))
plt.grid(True, which="both", linestyle='--', alpha=0.7)
plt.xticks(threads, labels=[f"{p}" for p in threads])

# speedup ideal
plt.plot(threads, threads, 'k--', label='Ideal', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#Eficiencia
threads = np.array([2, 4, 8, 16, 32, 64, 128])
Ns = np.array([32, 64, 128, 256, 512, 1024])
d = 8
C = 0.001
D = 0.0001

def T_parallel(N, p, d, C, D):
    epsilon = 10 / N
    work = (N**3) / (p )
    critical = C * (N**3) * np.log2(N) / (p )
    tasks = D * (8**d) * np.log2(N)
    return work + critical + tasks

def eficiencia(N, p, d, C, D):
    T1 = (N**3)
    Tp = T_parallel(N, p, d, C, D)
    return 1/(T1 / Tp)

plt.figure(figsize=(12, 7))
colors = itertools.cycle(plt.cm.tab10.colors)

for N, color in zip(Ns, colors):
    Es = [eficiencia(N, p, d, C, D) for p in threads]
    plt.plot(threads, Es, 'o-', label=f'N={N}', color=color, markersize=8)


plt.xscale('log', base=2)
plt.xlabel('Número de Threads (p)')
plt.ylabel('Eficiencia $E_p$')
plt.title('Eficiencia en Weak scaling')
plt.legend(title='Tamaño del cubo', bbox_to_anchor=(1.05, 1))
plt.grid(True, which="both", linestyle='--', alpha=0.7)
plt.xticks(threads, labels=[f"{p}" for p in threads])

# Línea de eficiencia ideal
plt.plot(threads, np.ones(7), 'k--', label='Ideal', alpha=0.5)
plt.tight_layout()
plt.show()
```
